```python
import os
import sys
import random
import numpy as np
import matplotlib.pylab as plt

sys.path.append('../.')
import utils
```

```python
OBJ_DIR = '/home/max/ieos/small_obj/vid_pred/data/seq_cls/human/train'
path_list = list(map(lambda x: os.path.join(OBJ_DIR, x), os.listdir(OBJ_DIR)))
```

```python
idx = random.randint(0, len(path_list)-1)
rand_path = path_list[idx]
sample, labels = utils.load_pickle(rand_path)


fig = plt.figure(figsize=(12,12))
row, cols = 5, 2
if labels == 1:
        label = 'human'
else:
    label = 'bg'
plt.title(label)
plt.axis('off')

for i in range(len(sample)):
    img  = sample[i]
    fig.add_subplot(row, cols, i+1)
    plt.title(str(img.shape[:2]))
    plt.axis('off')
    plt.imshow(img, cmap='gray')
plt.show()
```
