# Результаты обучения моделей последовательнрости на кропах людей

```python
import matplotlib.pylab as plt
import pandas as pd
```

```python
def plot_metrics_from_df(df, size=(16, 8)):
    epoches = df['epoch']
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=size)
    axes = axes.flatten()
    for i, metric in enumerate(df.columns[1:]):
        axes[i].plot(df[metric])
        axes[i].set_title(metric)
        axes[i].grid(True)
    plt.tight_layout()
    plt.show()
```
